# Import libraries

```python
import tensorflow as tf ### models
import matplotlib.pylab as plt ### ploting
```

# Load image

```python
# Load the original image
image_path = 'original_image.jpg'
#original_img_plt = load_img(image_path)
image_bytes = tf.io.read_file(image_path)
decoded_image = tf.image.decode_image(image_bytes)
original_image = tf.image.convert_image_dtype(decoded_image, tf.float32)
```

```python
original_image.shape
```

```python
type(original_image)
```

```python
original_image.shape
```

# visulaize method

```python
def visulaize(original_image, augmented_image):
    plt.subplot(1,2,1)
    plt.imshow(original_image)
    
    plt.subplot(1,2,2)
    plt.imshow(augmented_image)    
```

# tf.image.resize

```python
target_height = 224
target_width = 224
```

```python
resized_image = tf.image.resize_with_pad(original_image, target_height, target_width)
visulaize(original_image, resized_image)
```

# tf.image.resize_with_pad

### resize_with_pad resizes an image to a target width and height by keeping the aspect ratio the same without distortion. If the target dimensions don't match the image dimensions, the image is resized and then padded with zeroes to match requested dimensions.

```python
resized_image__with_pad =  tf.image.resize_with_pad(decoded_image, target_height, target_width)
visulaize(decoded_image, resized_image)
```

```python
resized_image__with_pad =  tf.image.resize_with_pad(original_image,target_height, target_width)
visulaize(original_image, resized_image__with_pad)
```

# tf.image.adjust_brightness

## delta should be in the range (-1,1)

```python
adjusted_image_brightness = tf.image.adjust_brightness(image = decoded_image, delta = 0.2)
visulaize(original_image, adjusted_image_brightness)
```

```python
adjusted_image_brightness = tf.image.adjust_brightness(decoded_image, -0.5)
visulaize(original_image, adjusted_image_brightness)
```

# tf.image.adjust_contrast

### contrast_factor must be in the interval (-inf, inf)

```python
adjusted_image_contrast = tf.image.adjust_contrast(decoded_image, 5.)
visulaize(original_image, adjusted_image_contrast)
```

```python
adjusted_image_contrast = tf.image.adjust_contrast(decoded_image, -5.)
visulaize(original_image, adjusted_image_contrast)
```
